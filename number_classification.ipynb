```python
import keras
from keras import layers
from keras.datasets import mnist
import numpy as np

(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255
test_images = test_images.reshape((10000, 28 * 28))
test_images = test_images.astype('float32') / 255

train_images_with_noise_channels = np.concatenate(
    [train_images, np.random.random((len(train_images), 784))],
    axis=1
)

test_images_with_noise_channels = np.concatenate(
    [test_images, np.random.random((len(test_images), 784))],
    axis=1
)

train_images_with_zeros_channels = np.concatenate(
    [train_images, np.zeros((len(train_images), 784))],
    axis=1
)

test_images_with_zeros_channels = np.concatenate(
    [test_images, np.zeros((len(test_images), 784))],
    axis=1
)

def get_model():
    model = keras.Sequential(
        [
            layers.Dense(512, activation='relu', input_shape=(784 * 2,)),
            layers.Dense(10, activation='softmax')
        ]
    )
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model


model = get_model()
history_noise = model.fit(
    train_images_with_noise_channels, 
    train_labels, 
    epochs=15, 
    batch_size=128,
    validation_split=0.2
)

# model = get_model()
# history_zeros = model.fit(
#     train_images_with_zeros_channels,
#     train_labels,
#     epochs=15,
#     batch_size=128,
#     validation_split=0.2
# )

# Draw the relationship between the number of epochs and the accuracy of the model, on training and validation data, 
# for both models (with noise channels and with zero channels). 
# You can use matplotlib to plot the accuracy curves.
import matplotlib.pyplot as plt

plt.plot(history_noise.history['accuracy'], label='Training Accuracy (Noise)')
plt.plot(history_noise.history['val_accuracy'], label='Validation Accuracy (Noise)')
# plt.plot(history_zeros.history['accuracy'], label='Training Accuracy (Zeros)')
# plt.plot(history_zeros.history['val_accuracy'], label='Validation Accuracy (Zeros)')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
